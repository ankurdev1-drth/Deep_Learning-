Question 14

```python
import torch.nn as nn
class Dense(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features)
        self.relu = nn.ReLU()

    def forward(self, X):
        return self.relu(self.linear(X))
```

```python
torch.manual_seed(42)
dense = Dense(3,5)
X = torch.randn(2,3)
y_pred = dense(X)
y_pred.shape
```
