### Regression MLPs-

-  MLPRegressor class from sklearn is used

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
fch = fetch_california_housing()
X_trainf, X_testf, y_trainf, y_test= train_test_split(fch.data, fch.target,  random_state=42)
X_train, X_valid, y_train, y_valid= train_test_split(X_trainf, y_trainf, random_state=42)
mlp_reg = MLPRegressor(hidden_layer_sizes=[50,50,50], random_state=42)
pipeline = make_pipeline(StandardScaler(), mlp_reg)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_valid)
rmse = root_mean_squared_error(y_valid, y_pred)
```

- MLP doesn't use any activation function for the output layer, so it's free to output any value it wants
    - MLP doesn't support actiavtion functions in the output layer
    - MLPRegressor only supports MSE
- to guarantee that the output will always be positive, ReLU activation function in the output layer must be used or the softplust activation function, which is smooth variant  of ReLU
- To gurantee that predictions always fall within a given range of values, then sigmoid function or the hyperbolic tangent should be use

### Classification MLPs

NOTE:
- for binary :
  - need a single output neuron using the sigmoid activation function, output will be between 0 and 1  which can interpret as the estimated probability of the positive class.
  - for negative class estimated probability is equal to 1 - that number
- easily handle multilabel binary classification tasks
-  minimizes the cross entropy rather than the MSE


```python
from sklearn.neural_network import MLPClassifier
mlpc = MLPClassifier(hidden_layer_sizes = [10, 10, 10], random_state=42)
pipeC = make_pipeline(StandardScaler(), mlpc)
```

### Learnings from exercises question @1

TensorFlow playground:
- Generally more the layers , more complex the patterns can be
- activation function plays and integral role in making the Nueral network fast/slow
- Risk of local minimum, when there are less hidden layers (for example 1 hidden layer with 3 neurons) then the training time varies a lot, and sometimes it even gets stuck in a local minimum
- When neural nets are too small  then it underfits the training set due to which it becomes incapable of finding a good solution
   - its anti-case what if it gets too  large? , large NN(neural networks) rarely get stuck in local minima, and even when they do these local optima are often almost as good as the global optimum. However, they can still get stuck on long plateaus for a long time!
- Risk of vanishing gradients in deep networks (i.e. basically in deep networks some neurons  tend to evolove faster as compared to other neurons present this is known as vanishing gradients problem)
