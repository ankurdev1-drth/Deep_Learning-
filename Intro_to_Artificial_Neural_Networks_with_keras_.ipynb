### Regression MLPs-

-  MLPRegressor class from sklearn is used

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
fch = fetch_california_housing()
X_trainf, X_testf, y_trainf, y_test= train_test_split(fch.data, fch.target,  random_state=42)
X_train, X_valid, y_train, y_valid= train_test_split(X_trainf, y_trainf, random_state=42)
mlp_reg = MLPRegressor(hidden_layer_sizes=[50,50,50], random_state=42)
pipeline = make_pipeline(StandardScaler(), mlp_reg)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_valid)
rmse = root_mean_squared_error(y_valid, y_pred)
```

```python
mlp_reg
```

```python
rmse
```

- MLP doesn't use any activation function for the output layer, so it's free to output any value it wants
    - MLP doesn't support actiavtion functions in the output layer
    - MLPRegressor only supports MSE
- to guarantee that the output will always be positive, ReLU activation function in the output layer must be used or the softplust activation function, which is smooth variant  of ReLU
- To gurantee that predictions always fall within a given range of values, then sigmoid function or the hyperbolic tangent should be use

### Learnings from exercises question @1

TensorFlow playground:
- Generally more the layers , more complex the patterns can be
- activation function plays and integral role in making the Nueral network fast/slow
- Risk of local minimum, when there are less hidden layers (for example 1 hidden layer with 3 neurons) then the training time varies a lot, and sometimes it even gets stuck in a local minimum
- When neural nets are too small  then it underfits the training set due to which it becomes incapable of finding a good solution
   - its anti-case what if it gets too  large? , large NN(neural networks) rarely get stuck in local minima, and even when they do these local optima are often almost as good as the global optimum. However, they can still get stuck on long plateaus for a long time!
- Risk of vanishing gradients in deep networks (i.e. basically in deep networks some neurons  tend to evolove faster as compared to other neurons present this is known as vanishing gradients problem)

### Building an Image Classifier Using the Sequential API

**Using Keras to load the dataset**

```python
import tensorflow as tf
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist
X_train, y_train = X_train_full[:-5000], y_train_full[:-5000]
X_valid, y_valid = X_train_full[-5000:], y_train_full[-5000:]
```

```python
X_train.shape
```

```python
X_train.dtype
```

```python
X_train, X_valid, X_test = X_train / 255.0, X_valid / 255.0, X_test / 255.
```

For simplicity we have scaled the pixel intensities down to the 0-1 range by dividign them with 255.0

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag",  "Ankle Boot"]
```

```python
class_names[y_train[0]]
```

# Creating the model using the sequential API

```python
tf.random.set_seed(42) #set Tensorflow's random seed to make the results reproducible
model = tf.keras.Sequential()  # Keras model for neural network that are just composed of a single stack of layers connected sequentially
model.add(tf.keras.layers.Input(shape=[28,28])) #first layers , keras need to know the shape of the input so it can determine the shape of the connection weight matrix of the first hidden layer
model.add(tf.keras.layers.Flatten()) #  Flatten layers, it converts each input image into a 1D array .if it receives input data X, it computes X.reshape.  this layer doesn't have any parameters
model.add(tf.keras.layers.Dense(300, activation="relu")) # Dense hidden layer with 300 neurons, it'll use the ReLU actiavtion function. also manages the  vector of bias terms
model.add(tf.keras.layers.Dense(100, activation="relu")) #second dense layer with 100 neurons,
model.add(tf.keras.layers.Dense(10, activation="softmax"))# output layer with 10 neurons (one per class), using the softmax activation function because the classes are exclusive
```

```python
model.summary()
```

### Compiling the model

Note:
- compile() method is called to specify the loss function and the optimizer to use
- a list of extra metrics can be passed for computing during training and evaluation

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

- 'sparse_categorical_crossentropy loss is used because we have sparse labels and classes are exclusive
- for binary classification or multilabel binary classification we use the "sigmoid" activation function in the output layer instead of the "softmax" activation function and "binary_crossentropy" loss
- tf.keras.utils.to_categorical function for conversion of sparse labels to  one-hot vector labels


### Training and evaluating the model

```python
history = model.fit(X_train, y_train, epochs=30, validation_data = (X_valid, y_valid))
```

Note:

- for per-instance weights, set the sample_weight argument
- for providing **sample weights** for the validation set can be done by adding them as a third item in the validation_data tuple

fit() method returns:

- history.history #dictionary containing loss and extra matrics it measured at the end of each epoch on the training set and on the validation set (if any)
- history.epoch # list of the epochs it went through

## Building a Regression MLP using the Sequential API

Note:
- Using the California housing problem
- output layer has a single neuron(since we want a single value as output)
- uses no activation function
- the loss function is mean squre error
- the metric is RMSE (root means square error)
- Using **Normalization** layers instead of **Flatten** as the first layer; does the same thing as sklearn's StandardScaler but adapt() method is used for  fitting it to the training set before calling fit() method

```python
import tensorflow as tf
tf.random.set_seed(42)
norm_layer = tf.keras.layers.Normalization(input_shape=X_train.shape[1:])
model = tf.keras.Sequential([
    norm_layer,
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(1)
])
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.layers
```

```python
h_3 = model.layers[1]
```

```python
h_3.dtype
```

```python
h_3
```

```python
h_3.get_weights()
```

```python
model.compile(loss="mse", metrics=["RootMeanSquaredError"], optimizer=optimizer)
norm_layer.adapt(X_train)
history = model.fit(X_train, y_train, epochs=20, validation_data=(X_valid, y_valid))
mse_test, rmse_test = model.evaluate(X_test, y_test)
X_new = X_test[:3]
y_pred = model.predict(X_new)
```

## Building Complex Models Using the Functional API

```python
tf.keras.backend.clear_session()
tf.random.set_seed(42)
```

Note:
- Functional API is used to build neural networks with more complex techonologies, or with multiple inputs/outputs
- Using the california housing problem

```python
normalization_layer = tf.keras.layers.Normalization()
hidden_layer_1 = tf.keras.layers.Dense(30, activation="relu")
hidden_layer_2 = tf.keras.layers.Dense(30, activation="relu")
concat_layer = tf.keras.layers.Concatenate()
output_layer = tf.keras.layers.Dense(1)

input_ = tf.keras.layers.Input(shape=X_train.shape[1:])
normalized = normalization_layer(input_)
hidden1 = hidden_layer_1(normalized)
hidden2 = hidden_layer_2(hidden1)
concat = concat_layer([normalized, hidden2])
output = output_layer(concat)
model = tf.keras.Model(inputs=[input_], outputs=[output])
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(loss="mse", optimizer=optimizer, metrics=["RootMeanSquaredError"])
normalization_layer.adapt(X_train)
history = model.fit(X_train, y_train, epochs=20,
                    validation_data=(X_valid, y_valid))
mse_test = model.evaluate(X_test, y_test)
y_pred = model.predict(X_new)
```

In case of multiple inputs

```python
input_wide = tf.keras.layers.Input(shape=[5]) #features 0 to 4
input_deep = tf.keras.layers.Input(shape=[6]) # features 2 to 7
norm_layer_wide = tf.keras.layers.Normalization()
norm_layer_deep = tf.keras.layers.Normalization()
norm_wide = norm_layer_wide(input_wide)
norm_deep = norm_layer_deep(input_deep)
hidden1 = tf.keras.layers.Dense(30, activation="relu")(norm_deep)
hidden2 = tf.keras.layers.Dense(30, activation="relu")(hidden1)
concat = tf.keras.layers.concatenate([norm_wide, hidden2])
output = tf.keras.layers.Dense(1)(concat)
model = tf.keras.Model(inputs=[input_wide, input_deep], outputs=[output])


```

```python
tf.keras.utils.plot_model(model)
```

```python
tf.keras.backend.clear_session()
tf.random.set_seed(42)

# Re-load California Housing data for this model
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
fch = fetch_california_housing()
X_trainf, X_test_cali, y_trainf, y_test_cali = train_test_split(fch.data, fch.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_trainf, y_trainf, random_state=42)

# Assign y_test to the California Housing target for evaluation
y_test = y_test_cali

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(loss="mse", optimizer=optimizer, metrics=["RootMeanSquaredError"])

X_train_wide, X_train_deep = X_train[:, :5], X_train[:, 2:]
X_valid_wide, X_valid_deep = X_valid[:, :5], X_valid[:, 2:]
X_test_wide, X_test_deep = X_test_cali[:, :5], X_test_cali[:, 2:] # Corrected: use X_test_cali
X_new_wide, X_new_deep = X_test_cali[:3, :5], X_test_cali[:3, 2:] # Corrected: use X_test_cali

norm_layer_wide.adapt(X_train_wide)
norm_layer_deep.adapt(X_train_deep)
history = model.fit((X_train_wide, X_train_deep), y_train, epochs=20,
                    validation_data=((X_valid_wide, X_valid_deep), y_valid))
mse_test = model.evaluate((X_test_wide, X_test_deep), y_test)
y_pred = model.predict((X_new_wide, X_new_deep))
```
